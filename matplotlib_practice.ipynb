```python
%matplotlib inline
```

```python
#import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create a line plot
plt.plot(x_axis,y_axis,label='Boston',linewidth=2,color='blue',marker='o')
#add title
plt.title("PyBer Fares by Month")
#set limits for the y-axis
plt.ylim(0,45)
#add labels for x-axis, and y-axis
plt.xlabel("Month")
plt.ylabel("Fare ($)")
#add legend
plt.legend()
#add a grid
plt.grid()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots(figsize=(10,8))
ax.plot(x_axis, y_axis,label='boston',color='green',marker='d',linewidth=2)
ax.set_title("Pyber Fares by Month")
ax.set_ylabel("Fare ($)")
ax.set_xlabel("Month")
ax.set_ylim(0,45)
ax.grid()
ax.legend()
plt.show()
```

```python
plt.bar(x_axis,y_axis,color="green",label='Boston')
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
plt.barh(x_axis,y_axis,color='magenta',label='Boston',alpha=0.7)
plt.gca().invert_yaxis()
plt.title("Pyber Fare by Month")
plt.ylabel("Months")
plt.xlabel("Fare ($)")
```

```python
fig,ax = plt.subplots()
ax.bar(x_axis,y_axis)
ax.set_title("Pyber Fare by Month")
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color="cyan",alpha=0.75,label="Chicago")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Month")
ax.set_title("Pyber Fare by Month")
ax.legend()
ax.invert_yaxis()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(x_axis,y_axis,s=[y*3 for y in y_axis])
```

```python
fig,ax = plt.subplots()
ax.scatter(y_axis,x_axis,s=[y*5 for y in y_axis],c='skyblue',edgecolor='grey',alpha=0.8,linewidth=2,label='Boston')
ax.set_title("Pyber Fares by Month")
ax.legend()
ax.set_xlim(0,max(y_axis)+5)
ax.invert_yaxis()
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%',colors=colors)
```
